Use spaCy to part-of-speech tag a sentence. Use the trained model **en_core_web_sm**

```python
import spacy
nlp = spacy.load("en_core_web_sm",disable=['parser'])
```

Pass in the sentence *The mouse eats the cheese* to be tagged.

```python
sentence = nlp('The mouse eats the cheese.')
```

Iterate through the sentence, printing the word and its part-of-speech tag. The tagset that is used is the [Universal part-of-speech tagset](https://universaldependencies.org/u/pos/).

```python
for word in sentence:
  print(word.text, word.pos_)
```

Iterate through the sentence again, printing the word and its part-of-speech tag. This time, the tagset that is printed is the more fine-grained [Penn Treebank tagset](https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html).


```python
for word in sentence:
  print(word.text,word.tag_)
```

Print the explanations for some of the Penn Treebank tags.

```python
from spacy import explain
def explain_tags(tags):
  for tag in tags:
    print(tag, explain(tag))

explain_tags(['DT','NN','VBZ'])
```
