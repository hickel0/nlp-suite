```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

Pass in the sentence *Kamala Harris debated against Donald Trump in the National Constitution Center in Philadephia.*

```python
sentence = nlp('Kamala Harris debated against Donald Trump in the National Constitution Center in Philadephia.')
```

Print out the named entities found in the sentence, along with their type.

```python
for ent in sentence.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import explain
def explain_tags(tags):
  for tag in tags:
    print(tag, explain(tag))

explain_tags(['ORG','GPE'])
```
